```python
cities = cities[['year','wikipedia_density', 'city', 'wikipedia_area', 'wikipedia_creation_date', 'wikipedia_city_rights_since', 'wikipedia_nr_of_people']]
```

```python
cities['wikipedia_area'] = cities['wikipedia_area'].round(0).astype(int)
cities['wikipedia_nr_of_people'] = cities['wikipedia_nr_of_people'].round(0).astype(int)
cities['wikipedia_density'] = cities['wikipedia_density'].round(0).astype(int)
```

```python
cities.head()
```

```python
cities['wikipedia_area'] = (cities['wikipedia_area']/10).astype(int)*10
cities['wikipedia_nr_of_people'] = (cities['wikipedia_nr_of_people']/100).astype(int)*100
cities['wikipedia_density'] = (cities['wikipedia_density']/100).astype(int)*100
```

```python
miasta_z_wiki = miasta_z_wiki.rename(columns={"area": "wikipedia_area", "people": "wikipedia_nr_of_people", "density":"wikipedia_density"})
```

```python
miasta_z_wiki['wikipedia_area'] = (miasta_z_wiki['wikipedia_area']/10).astype(int)*10
miasta_z_wiki['wikipedia_nr_of_people'] = (miasta_z_wiki['wikipedia_nr_of_people']/100).astype(int)*100
miasta_z_wiki['wikipedia_density'] = (miasta_z_wiki['wikipedia_density']/100).astype(int)*100
```

```python
cities_merge = cities.merge(miasta_z_wiki, left_on=['wikipedia_nr_of_people'], right_on=['wikipedia_nr_of_people'])
```
